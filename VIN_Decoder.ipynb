```python
import pandas as pd
```

```python
# VIN input and data collection
vin_file = open('VIN.csv') # opening the file containing VINs, name the file as "VIN.csv" include all the VINs in the first column, no column header, save it in the same directory as this Notebook 

all_attr_df = pd.DataFrame()

for vin in vin_file:
    url = ('https://vpic.nhtsa.dot.gov/api/vehicles/decodevinvalues/{}?format=json'.format(vin.strip()))
    full_json = pd.read_json(url) # reading the full info in a JSON format from NHTSA
    full_attr_dict = full_json.Results[0] # reading full vehicle attributes from the read JSON file. The read data is in the dictionary format
    full_attr_df = pd.DataFrame.from_dict(full_attr_dict, orient='index').transpose() # converting the dictionary to a dataframe and transposing it
    all_attr_df = all_attr_df.append(full_attr_df) # append all data to one dataframe
```

```python
# output
required_columns = ['VIN', 'Make', 'Model', 'ModelYear', 'BodyClass', 'ElectrificationLevel',
                    'ErrorCode', 'FuelTypePrimary', 'FuelTypeSecondary', 'VehicleType'] # creatig a list of required columns, modify it based on your needs

all_attr_df[required_columns].set_index('VIN').to_csv('VIN_pulled_data.csv') # save the results into a csv file
```
